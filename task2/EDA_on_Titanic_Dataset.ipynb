```python
# Step 3: Create Visualizations Using Matplotlib and Seaborn
# 3.1 Distribution of Age
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'].dropna(), bins=30, kde=True)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# 3.2 Survival Count by Gender
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Sex', hue='Survived')
plt.title('Survival Count Based on Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Survived', loc='upper right', labels=['No', 'Yes'])
plt.show()
```

```python
# 3.3 Survival Rate by Class
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='Pclass', y='Survived', ci=None)
plt.title('Survival Rate by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```
